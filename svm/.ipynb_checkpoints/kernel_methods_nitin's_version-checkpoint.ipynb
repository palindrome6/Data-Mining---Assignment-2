```python
X, y = make_blobs(n_samples=1000, centers=2, n_features=2,
                      random_state=0)
```

<p style="color: #f03b20">Train an SVM on the dataset using respectively a linear, polynomial and radial basis function (RBF)
kernel, evaluate the performance of each kernel using 10-fold cross-validation and AUC (area under

First, we train an SVM on the data set using the three kernels.

```python
svc = svm.SVC(kernel='linear').fit(X, y)
rbf_svc = svm.SVC(kernel='rbf').fit(X, y)
poly_svc = svm.SVC(kernel='poly').fit(X, y)
```

The performance of the three kernels are evaluated using crossfold validation and AUC.

```python
print "Mean auc score for linear kernel:"
scores = cross_validation.cross_val_score(svc, X, y, cv=10, scoring='roc_auc')
print scores.mean()
print "Mean auc score for rbf kernel:"
scores = cross_validation.cross_val_score(rbf_svc, X, y, cv=10, scoring='roc_auc')
print scores.mean()
print "Mean auc score for polynomial kernel:"
scores = cross_validation.cross_val_score(poly_svc, X, y, cv=10, scoring='roc_auc')
print scores.mean()
```

The results are visualized.

```python
h = .02 
plt.rcParams["figure.figsize"] = (16, 4)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# title for the plots
titles = ['SVC with linear kernel',
          'SVC with RBF kernel',
          'SVC with polynomial kernel']


for i, clf in enumerate((svc, rbf_svc, poly_svc)):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    plt.subplot(1, 3, i + 1)
    plt.subplots_adjust(wspace=0.5, hspace=0.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('x coordinate')
    plt.ylabel('y coordinate')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])

plt.show()
```

```python
clf1 = svm.SVC(kernel='rbf',C=math.pow(2,-4)).fit(X, y)
clf2 = svm.SVC(kernel='rbf',C=math.pow(2,-2)).fit(X, y)
clf3 = svm.SVC(kernel='rbf',C=math.pow(2,0)).fit(X, y)
clf4 = svm.SVC(kernel='rbf',C=math.pow(2,2)).fit(X, y)
clf5 = svm.SVC(kernel='rbf',C=math.pow(2,4)).fit(X, y)
clf6 = svm.SVC(kernel='rbf',gamma=math.pow(2,-4)).fit(X, y)
clf7 = svm.SVC(kernel='rbf',gamma=math.pow(2,-2)).fit(X, y)
clf8 = svm.SVC(kernel='rbf',gamma=math.pow(2,0)).fit(X, y)
clf9 = svm.SVC(kernel='rbf',gamma=math.pow(2,2)).fit(X, y)
clf10 = svm.SVC(kernel='rbf',gamma=math.pow(2,4)).fit(X, y)
```

C controls the cost of misclassification and determines whether the margin is a soft-margin or a hard-margin.

A small C value gives a low misclassificaiton cost (soft margin),thus leading to a higher possibility of misclassifications in search for a wider cushion.

A large C value gives for a hard margin and thus could potentially overfit the data.

From the scores, however, this can not be clearly observed.


In general, it can be observed that the value of gamma has a profound effect of the scoring. In general, it can be observed that s the value of gamma increases, the score (AUC) reduces. 

For a gamma value of 2^6, we can see clearly that there is overfitting and hence this is not desirable.

As the vallue of gamma reaches 2^15,  we can observe here that all the points are judged to be in the same class. Hence there is absolute overfitting. 


Hence, a large gamma value gives high bias even though the variance is comparatively low.

<p style="color: #f03b20">Visualize the results in a plot Cxgamma -> p with p being performance of the model (AUC) visualized
as the color of the data point. If you don't know how to do this, at least plot C -> p  for various values
of gamma.</p>

```python
# Utility function to move the midpoint of a colormap to be around the values of interest.

class MidpointNormalize(Normalize):

    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```

```python
C_range = np.logspace(-15, 15, 10, base=2)
gamma_range = np.logspace(-15, 15, 10, base=2)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = StratifiedShuffleSplit(y, n_iter=5, test_size=0.2, random_state=42)
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv)
grid.fit(X, y)

print("The best parameters are %s with a score of %0.2f"
      % (grid.best_params_, grid.best_score_))

scores = [x[1] for x in grid.grid_scores_]
scores = np.array(scores).reshape(len(C_range), len(gamma_range))
```

```python
plt.rcParams["figure.figsize"] = (20, 8)
plt.style.use('classic')
plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(scores, interpolation='nearest', cmap=get_cmap("YlOrBr"),
           norm=MidpointNormalize(vmin=0.2, midpoint=0.92))
plt.xlabel('gamma')
plt.ylabel('C')
plt.colorbar()
plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=90)
plt.yticks(np.arange(len(C_range)), C_range)
plt.title('Accuracy')
plt.show()
```
